# Verspätungs-mean für Haltestellen berechnen 

```python
def fromiso(dstr):
    datetime_object = datetime.strptime(dstr, '%Y-%m-%dT%H:%M:%SZ')
    return datetime_object
```

```python
import sys
```

```python
stops = {}
with open(f'../../../fahrten_jsons/12859858.json', 'r') as json_data:
        data = json.load(json_data)["stops"]
        #print(data)
        for stop in data:
            try:
                delta = fromiso(stop['properties']['abfahrtprognose']) - fromiso(stop['properties']['abfahrt'])
                print(delta)
                try:      
                    stops[stop['properties']['halteid']].append(delta)
                except:
                    stops[stop['properties']['halteid']] = [delta]
                    
            except ValueError as e:
                print(e)
            
            
        
        
        
```

```python
stops = {}
for root, dirs, files in os.walk('../../../fahrten_jsons/'):
    for file in files:
        path = os.path.join(root, file)
        with open(path, 'r') as json_data:
            data = json.load(json_data)["stops"]
            #print(data)
            for stop in data:
                try:
                    print(stop['properties']['abfahrtprognose'])
                    delta = fromiso(stop['properties']['abfahrtprognose']) - fromiso(stop['properties']['abfahrt'])
                    print(delta)
                    try:      
                        stops[stop['properties']['halteid']]['deltas'].append(delta)
                        stops[stop['properties']['halteid']]['coordinates'] = stop['geometry']['coordinates']
                    except:
                        stops[stop['properties']['halteid']] = {}
                        stops[stop['properties']['halteid']]['deltas'] = [delta]
                        stops[stop['properties']['halteid']]['coordinates'] = stop['geometry']['coordinates']

                except ValueError as e:
                    #print(e)
                    continue
        
        
```

## Average delay

```python
stop_means = {}
for key,val in stops.items():
    transformed_values = [delta.total_seconds() for delta in val['deltas']]
    stop_means[key] = {}
    stop_means[key]['mean'] = np.mean(np.asarray(transformed_values))
    stop_means[key]['coordinates'] = val['coordinates']
```

```python
with open(f'../../../fahrten_jsons/average_delay.json', 'w') as f:
        json.dump(stop_means, f)
```

## Maximal delay

```python
stop_maximum = {}
for key,val in stops.items():
    transformed_values = [delta.total_seconds() for delta in val['deltas']]
    stop_maximum[key] = np.max(np.asarray(transformed_values))
```

```python
with open(f'../../../fahrten_jsons/maximal_delay.json', 'w') as f:
        json.dump(stop_maximum, f)
```

```python
max_delay = np.max([v['mean'] for _, v in stop_means.items()])
max_delay

heatmap_data = {
   'meanDelays': stop_means,
   'maxDelay': max_delay
}

json.dump(heatmap_data, open(f'../../../fahrten_jsons/heatmap_data.json', 'w'))
```
